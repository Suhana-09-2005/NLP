### Explanation of NLTK POS Tags:

NLTK uses the Penn Treebank tagset for POS tagging. Here are a few common tags you might see:

*   **NN**: Noun, singular or mass (e.g., `research`, `impact`)
*   **NNS**: Noun, plural (e.g., `protocols`)
*   **NNP**: Proper noun, singular (e.g., `Quantum`)
*   **JJ**: Adjective (e.g., `profound`, `cryptographic`)
*   **VBZ**: Verb, 3rd person singular present (e.g., `investigates`)
*   **IN**: Preposition or subordinating conjunction (e.g., `of`, `on`)
*   **DT**: Determiner (e.g., `This`, `the`)
*   **CC**: Coordinating conjunction (e.g., `and`)

You can find a full list of Penn Treebank POS tags and their meanings in the NLTK documentation or online resources.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a startup in India.")
for token in doc:
  print(token.text, token.pos_, token.tag_)
```

```python
str=""
doc = nlp(str)
words = nltk.word_tokenize(str)
pos_tags = nltk.pos_tag(words)
spacy_tags=[]
for token in doc:
    spacy_tags.append((token.text, token.pos_, token.tag_))

print("NLTK tags                  Spacy tags")
for i in range(len(pos_tags)):
    print(f"{pos_tags[i]}          {spacy_tags[i]}")
nltk_noun=0
nltk_verb=0
spacy_noun=0
spacy_verb=0
for (i,j) in pos_tags:
    if j.startswith('V'):
        nltk_verb= nltk_verb+1
    if j.startswith('N'):
        nltk_noun=nltk_noun+1
for i,j,k in spacy_tags:
    if k.startswith('V'):
        spacy_verb= spacy_verb+1
    if k.startswith('N'):
        spacy_noun=spacy_noun+1
print("NLTK Noun count:",nltk_noun)
print("Spacy Noun count:",spacy_noun)
print("NLTK Verb count:",nltk_verb)
print("Spacy Verb count:",spacy_verb)
```
