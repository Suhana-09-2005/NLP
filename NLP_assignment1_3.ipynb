```python
paragraph = """The quick brown fox jumps over the lazy dog. This is a sample paragraph for text preprocessing demonstration. We will count its words, convert it to lowercase, and then split it."""

# 1. Count the number of words in the text
words = paragraph.split()
word_count = len(words)
print(f"Original paragraph: {paragraph}")
print(f"Number of words: {word_count}")

# 2. Convert the entire text to lowercase
lowercase_paragraph = paragraph.lower()
print(f"\nLowercase paragraph: {lowercase_paragraph}")

# 3. Split the lowercase text into words
split_words = lowercase_paragraph.split()
print(f"\nSplit words (from lowercase text): {split_words}")
```
